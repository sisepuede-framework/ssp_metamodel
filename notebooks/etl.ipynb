```python
import pandas as pd
import os
```

```python
SCRIPT_DIR_PATH = os.getcwd()
ROOT_DIR_PATH = os.path.dirname(SCRIPT_DIR_PATH)
DATA_DIR_PATH = os.path.join(ROOT_DIR_PATH, "data")
MAPPING_DIR_PATH = os.path.join(DATA_DIR_PATH, "mapping")
PREDICTORS_DIR_PATH = os.path.join(DATA_DIR_PATH, "predictors")
SSP_DIR_PATH = os.path.join(DATA_DIR_PATH, "ssp")
SIMULATION_DIR_PATH = os.path.join(SSP_DIR_PATH, "sisepuede_run_2025-01-14T17;04;06.975301_output_database")
TRAINING_DIR_PATH = os.path.join(DATA_DIR_PATH, "training")
```

```python
lhc_trials_df = pd.read_csv(os.path.join(PREDICTORS_DIR_PATH, "lhs_sample_group_experiments.csv"))
lhc_trials_df
```

```python
var_specifications_df = pd.read_csv(os.path.join(MAPPING_DIR_PATH, "variable_specification_to_sample_group.csv"))
var_specifications_df
```

```python
var_specifications_df["sample_group"].unique().shape
```

```python
la_emissions_df = pd.read_csv(os.path.join(SIMULATION_DIR_PATH, "MODEL_OUTPUT.csv"))
la_emissions_df
```

```python
subsector_totals_cols = [col for col in la_emissions_df.columns if "emission_co2e_subsector_total" in col]

# Calculate total_emissions
la_emissions_df["total_emissions"] = la_emissions_df[subsector_totals_cols].sum(axis=1)

cols_to_keep = ["primary_id", "time_period", "total_emissions"]

la_emissions_df = la_emissions_df[cols_to_keep]

la_emissions_df
```

```python
# Filter out rows with time_period <= 30
la_emissions_df = la_emissions_df[la_emissions_df["time_period"] > 30]

# aggregate data by primary_id and region by summing the emissions
la_emissions_df_agg = la_emissions_df.groupby(["primary_id"]).sum().reset_index()

# Rename the total_emissions column to total_emissions_last_five_years
la_emissions_df_agg.rename(columns={"total_emissions": "total_emissions_last_five_years"}, inplace=True)

la_emissions_df_agg
```

```python
attr_primary_df = pd.read_csv(os.path.join(SIMULATION_DIR_PATH, "ATTRIBUTE_PRIMARY.csv"))
attr_primary_df
```

```python
la_emissions_df_w_future_id = la_emissions_df_agg.merge(attr_primary_df, on="primary_id", how="inner")

# Drop design_id and stratgy_id columns
la_emissions_df_w_future_id = la_emissions_df_w_future_id.drop(columns=["design_id", "strategy_id", "time_period"])
la_emissions_df_w_future_id
```

```python
merged_df = la_emissions_df_w_future_id.merge(lhc_trials_df, on="future_id", how="inner")

# Rearrange columns by moving "total_emissions_last_five_years" to the back
cols = list(merged_df.columns)
cols.remove("total_emissions_last_five_years")
cols.append("total_emissions_last_five_years")
merged_df = merged_df[cols]


merged_df
```

```python
var_traj_df = pd.read_csv(os.path.join(MAPPING_DIR_PATH, "var_trajgroups_experiment.csv"))

# missing_var_groups = [11, 21, 32, 33, 41, 42]
missing_var_groups = [11]

# filer var_traj_df to only include rows that are not  in missing_var_groups
var_traj_df = var_traj_df[~var_traj_df["variable_trajectory_group"].isin(missing_var_groups)]
var_traj_df.tail(10)
```

```python
relevant_lhs_cols = var_traj_df["variable_trajectory_group"].values

relevant_lhs_cols = [str(col) for col in relevant_lhs_cols.tolist()]
relevant_lhs_cols
```

```python
#filer merged_df to keep only relevant columns
merged_df_filtered = merged_df[["primary_id", "future_id"] + list(relevant_lhs_cols) + ["total_emissions_last_five_years"]]
```
